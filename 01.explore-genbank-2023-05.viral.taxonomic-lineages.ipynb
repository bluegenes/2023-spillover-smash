## add lineage column to gb taxonomy

```python

rank_columns = [
        'superkingdom', 'realm', 'subrealm', 'kingdom',
        'subkingdom', 'phylum', 'subphylum', 'class', 'subclass', 'order',
        'suborder', 'family', 'subfamily', 'genus', 'subgenus', 'species'
    ]

# Combine the columns with ';' as separator, filling NaNs with empty string
gb['lineage'] = gb[rank_columns].fillna('').apply(lambda row: ';'.join(row.values.astype(str)), axis=1)
```
